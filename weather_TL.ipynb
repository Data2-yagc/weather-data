```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
filename = "antananarivo_2020-11-17_to_2023-08-13.csv"
```

```python
df = pd.read_csv(f"./data/{filename}")
```

```python
columns = ["humidity", "precip", "windgust", "winddir", "tempmax", "tempmin", "windspeed", "datetime"]
df = pd.DataFrame(df.loc[:, columns]).fillna(0)
df["datetime"] = pd.to_datetime(df["datetime"])
df["year"] = df.datetime.dt.year
df["month"] = df.datetime.dt.month

data2020 = df.where(df["year"] == 2020)
data2021 = df.where(df["year"] == 2021)
data2022 = df[df["year"] == 2022]
data2023 = df[df["year"] == 2023]
```

```python
df.to_csv(f"./result/{filename}")
```

```python
x = "month"
y = "precip"
```

```python
plt.xlabel(x)
plt.ylabel(y)
sns.lineplot(data=data2020, x=x, y=y)
```

```python
plt.xlabel(x)
plt.ylabel(y)
sns.lineplot(data=data2021, x=x, y=y)
```

```python
plt.xlabel(x)
plt.ylabel(y)
sns.lineplot(data=data2022, x=x, y=y)
```

```python
plt.xlabel(x)
plt.ylabel(y)
sns.lineplot(data=data2023, x=x, y=y)
```
